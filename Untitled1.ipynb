```python
#Librerias
import numpy as np                
import cv2 as cv                  
import matplotlib.pyplot as plt   
from PIL import Image             
```

```python
#Preprocesado 
img1 = Image.open('/content/drive/MyDrive/Colab Notebooks/data/BeFunky-photo')          # queryImage
img1 = img1.resize((640,320), Image.ANTIALIAS)
img1 = np.array(img1)

img2 = Image.open('/content/drive/MyDrive/Colab Notebooks/data/test2.jfif') # trainImage
img2 = img2.resize((640,320), Image.ANTIALIAS)
img2 = np.array(img2)

b,g,r = cv.split(img1);
clahe = cv.createCLAHE(clipLimit=4.0, tileGridSize=(16,16))
b = clahe.apply(b)
g = clahe.apply(g)
r = clahe.apply(r)
bgr = cv.merge([b,g,r]);
img1 = cv.cvtColor(bgr,cv. COLOR_BGR2RGB);
```

```python
#ORB descriptor
orb = cv.ORB_create()
#Encuentra los descriptores con ORB
kp1, des1 = orb.detectAndCompute(img1,None)
kp2, des2 = orb.detectAndCompute(img2,None)
#Encuentra los descriptores comunes entre ambas imagenes
bf = cv.BFMatcher()
matches = bf.knnMatch(des1,des2,k=2)
#Distancia entre descriptores tolerada
good = []
for m,n in matches:
    if m.distance < 0.75*n.distance:
        good.append([m])
#impresion del resultado
img3 = cv.drawMatchesKnn(img1,kp1,img2,kp2,good,None,flags=cv.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
plt.imshow(img3,aspect="auto"),plt.show()
```
